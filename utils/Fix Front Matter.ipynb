```python
%matplotlib inline
import pandas as pd
import os, glob
import frontmatter
import random

import json
import pprint
pretty = pprint.PrettyPrinter(indent=4)
```

```python
# ar_Images = []

for file in glob.glob('../source/_drafts/**/*.md', recursive=True):
    with open(file, 'r+') as f:
        post = frontmatter.load(f)
        
        #print(post.keys())
            
        # Check for photo paths = ensure we have '/images/' not '/assets/' or 'images/'
        if 'photos' in post.metadata:
            for idx, image_path in enumerate(post.metadata.get('photos')):

                # But dont double up the slashes
                if image_path.startswith('images/'):
                    post.metadata['photos'][idx] = image_path.replace("images/", "/images/")
                    
                # fix photos that are referencing 'assets'
                if 'assets' in image_path:
                    # print(idx, image_path)   
                    # str.replace(old, new[, max])
                    post.metadata['photos'][idx] = image_path.replace("/assets/", "/images/")
        
        # Set Defaults
        if ('description' not in post.metadata) or (post.metadata['description'] == None):
            post.metadata['description'] = 'description - lorem ipsum'
            
        if 'subtitle' not in post.metadata or (post.metadata['subtitle'] == None):
            post.metadata['subtitle'] = 'subtitle - lorem ipsum'

        if ('comments' not in post.metadata) or (post.metadata['comments'] == None):
            post.metadata['comments'] = False

        if ('toc' not in post.metadata) or (post.metadata['toc'] == None):
            post.metadata['toc'] = True    
        
        # Default + Lower case what is there
        if 'categories' not in post.metadata or (post.metadata['categories'] == None):
            post.metadata['categories'] = ['uncategorized']
        else:
             post.metadata['categories'] = [cat.lower() for cat in post.metadata['categories']]

        if 'tags' in post.metadata:
            post.metadata['tags'] = [tag.lower() for tag in post.metadata['tags']]
        
        # Save obj back to file
        # print('updated: ' + file)
        # print(frontmatter.dumps(post))
        
        pretty.pprint(post.metadata)
        print('')
        print('')
        
        # post includes the body/contents thus dumps puts it all back and then lops off the file right after the insertion.
        f.seek(0)
        f.write(frontmatter.dumps(post))
        f.truncate()
```
